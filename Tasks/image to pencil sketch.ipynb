### 01 Beginner Level Task: Image to pencil sketch with python 

Procedure: 
   1. Import the libraries
   2. Load and the original image
   3. Convert the image tro grayscale image
   4. Check the minimun and maximum value of pixels
   5. Invert the image
   6. Blur the image
   7. Pencil sketch the image
   8. Save the pencil sketched image

#### Importing the image

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

%matplotlib inline
```

#### Loading and reading the original image

```python
image = cv.imread('./data/images.jpg')
plt.figure(figsize = (7, 8))
plt.title("Original Image")
plt.imshow(image)
plt.show()
```

#### Converting the image to a grayscale image

```python
g_image = cv.imread('./data/images.jpg', 0)
plt.figure(figsize = (7, 8))
plt.title("Grayscale Image")
plt.imshow(g_image, cmap = 'gray')
plt.show()
```

#### Checking the minimum and maximum values of pixels in the original image 

```python
min_value = image.min()
max_value = image.max()

print(f"The minimum value of pixels present in the image is {min_value}")
print(f"The maximum value of pixels present in the image is {max_value}")
```

#### Inverting the image

```python
i_image = 255 - g_image
plt.figure(figsize = (7, 8))
plt.title("Inverted Image")
plt.imshow(i_image, cmap = 'gray')
plt.show()
```

#### Blurring the image

```python
b_image = cv.GaussianBlur(i_image, ksize = (15, 15), sigmaX = 0, sigmaY = 0)
plt.figure(figsize = (7, 8))
plt.title("Blurred Image")
plt.imshow(b_image, cmap = 'gray')
plt.show()
```

#### Pencil sketch the image

```python
def pencil_sketch(x, y):
    return cv.divide(x, 255 - y, scale = 250)
```

```python
p_image = pencil_sketch(g_image, b_image)
plt.figure(figsize = (7, 8))
plt.title("pencil sketch Image")
plt.imshow(p_image, cmap = 'gray')
plt.show()
```

#### Saving the pencil sketched immage to a local directory

```python
cv.imwrite('./data/pencil sketched images.jpg', p_image)
```

The pencil sketch image has been saved to a local directory
